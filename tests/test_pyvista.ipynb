```python
import pyvista as pv

# Jupyter Notebook用のバックエンド設定（重要）
pv.set_jupyter_backend('trame')

def check_pyvista_on_ipynb():
    # 1. テスト用のメッシュ作成（UUVに見立てた円柱）
    uuv_test = pv.Cylinder(center=(0, 0, 0), direction=(1, 0, 0), radius=0.5, height=3.0)
    
    # 2. プロッターの作成
    plotter = pv.Plotter(notebook=True, window_size=[600, 400])
    
    # 3. メッシュの追加
    plotter.add_mesh(uuv_test, color='cyan', show_edges=True, label='UUV Model')
    
    # 4. 視覚効果の追加
    plotter.add_axes()
    plotter.show_grid()
    plotter.add_legend()
    
    # 5. 表示
    return plotter.show()

# 実行
check_pyvista_on_ipynb()
```
