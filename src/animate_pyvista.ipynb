```python
import pandas as pd
import pyvista as pv
import os

# パス設定（環境に合わせて調整してください）
# ipynbファイルから見て ../data/uuv_initial_path.csv にある想定
csv_path = '../data/uuv_trajectory.csv' 

if os.path.exists(csv_path):
    # 1. データの読み込み
    df = pd.read_csv(csv_path)
    points = df[['x', 'y', 'z']].values

    # 2. プロッターの準備
    pv.set_jupyter_backend('trame')
    plotter = pv.Plotter()

    # 3. 軌跡（パス）を線として追加
    # 深度(z)を反転させて表示したい場合は points[:, 2] = -points[:, 2] などに調整
    path = pv.MultipleLines(points)
    plotter.add_mesh(path, color='yellow', line_width=3, label='UUV Trajectory')

    # 4. 現在地点（最後）にUUVに見立てた矢印を配置
    last_pos = points[-1]
    # 進行方向（Yaw）を簡易的に反映させるための矢印
    uuv_model = pv.Arrow(start=last_pos, direction=(1, 0, 0), scale=2.0)
    plotter.add_mesh(uuv_model, color='cyan', label='Current UUV')

    # 5. 環境設定
    plotter.add_axes()
    plotter.show_grid()
    plotter.add_legend()
    
    # 表示
    plotter.show()
else:
    print(f"CSVファイルが見つかりません: {csv_path}")
```
